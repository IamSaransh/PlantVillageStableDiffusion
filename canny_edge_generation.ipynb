```python
import cv2
import os
from tqdm import tqdm
import cv2
import matplotlib.pyplot as plt

input_dir = "C:\workspace\Research\Diffusion\PlantAugmentation\images_all_in_one_folder"
output_dir = "/kaggle/working/sampele1"
os.makedirs(output_dir, exist_ok=True)
```

```python
i = 0
for img_name in tqdm(os.listdir(input_dir)):
    i = i+1
    img_path = os.path.join(input_dir, img_name)
    print(img_path)
    
    image_original = (cv2.imread(img_path))
    # Read the image in grayscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
    img = cv2.GaussianBlur(img, (5, 5), 0)
    
    # Apply Canny edge detection
    img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize=61, C=15)
    edges = cv2.Canny(img, threshold1=50, threshold2=170)  # Adjust thresholds as needed
    
    # Display the original and edge-detected images side by side
    plt.figure(figsize=(12, 4))  # Set the figure size
    
    # Original image
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
    plt.imshow(image_original)
    plt.title("Original Image")
    plt.axis("off")
    
    # Canny edge-detected image
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
    plt.imshow(edges, cmap="gray")
    plt.title("Canny Edge Detection")
    plt


    if(i==10):
        break
```
